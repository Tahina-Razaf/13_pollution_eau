```python
#view table mapping_categories verification des  cdparametresiseeaux non unique
query_map_unique = """
WITH duplicate_count AS (
SELECT cdparametre, cdparametresiseeaux, categorie, COUNT( cdparametresiseeaux) AS count_param
FROM mapping_categories
GROUP BY cdparametresiseeaux, cdparametre, categorie )
,
param_sup2 AS (SELECT * FROM duplicate_count
WHERE count_param > 1
ORDER BY cdparametresiseeaux)

SELECT *
FROM mapping_categories
WHERE cdparametresiseeaux IN (SELECT cdparametresiseeaux
                                FROM param_sup2)
ORDER BY cdparametresiseeaux

"""

table_check_dupliate_show = con.sql(query_map_unique).df()
table_check_dupliate_show

# cdparametre contient 61 lignes vides
# cdparametresiseeaux contient 1 ligne vide
```

```python
#join edc_resultat avec mapping_categories avec distinct cdparametresiseau
query_mapped_edc_resultats = """
SELECT edc_resultats.*, mapping.categorie
FROM edc_resultats
LEFT JOIN (SELECT DISTINCT cdparametresiseeaux, categorie FROM mapping_categories) AS mapping
ON edc_resultats.cdparametresiseeaux = mapping.cdparametresiseeaux
LIMIT 5
"""
mapped_edc_resultats_show = con.sql(query_mapped_edc_resultats).fetchdf()
mapped_edc_resultats_show

```

```python
#verif nombre de ligne après mapping entre edc_resultats et mapping_categories
count_mapped_edc_resultats = """
WITH MAPPED_EDC_RESULTATS AS (SELECT edc_resultats.*, mapping.categorie
FROM edc_resultats
LEFT JOIN (SELECT DISTINCT cdparametresiseeaux, categorie FROM mapping_categories) AS mapping
ON edc_resultats.cdparametresiseeaux = mapping.cdparametresiseeaux)

SELECT 
    CASE 
        WHEN Count(*)  = (SELECT count(*) FROM mapped_edc_resultats)
        THEN 'OK' 
        ELSE 'KO'
    END AS Comparaison
FROM edc_resultats

"""
nb_mapped_edc_resultats = con.sql(count_mapped_edc_resultats)
nb_mapped_edc_resultats

```

```python
#nombre de résultats par prélévements
query_resultat_parcatégorie_par_pvl = """
WITH MAPPED_EDC_RESULTATS AS (
SELECT edc_resultats.*, mapping.categorie
FROM edc_resultats
LEFT JOIN (SELECT DISTINCT cdparametresiseeaux, categorie FROM mapping_categories) AS mapping
ON edc_resultats.cdparametresiseeaux = mapping.cdparametresiseeaux)
,
LAST_PREL AS ( 
SELECT cdreseau, referenceprel, MAX(dateprel) AS dernierdateplv, plvconformitebacterio, plvconformitechimique, 
        plvconformitereferencebact, plvconformitereferencechim
FROM edc_prelevements
WHERE cdreseauamont IS NULL
GROUP BY cdreseau, referenceprel, plvconformitebacterio, plvconformitechimique, 
        plvconformitereferencebact, plvconformitereferencechim
ORDER BY referenceprel, dernierdateplv)
,
TEMP_RESULTS AS ( 
SELECT referenceprel, categorie, referenceanl, cdparametresiseeaux,rqana, limitequal, refqual, valtraduite 
FROM MAPPED_EDC_RESULTATS
ORDER BY referenceprel, categorie )
,
NB_RESULT_PAR_PLV AS (
SELECT referenceprel, COUNT(DISTINCT categorie) AS nbcategorieparplv
FROM TEMP_RESULTS
GROUP BY referenceprel
ORDER BY nbcategorieparplv DESC)

SELECT *
FROM NB_RESULT_PAR_PLV
LIMIT 10
"""

nb_resultats_par_plv = con.sql(query_resultat_parcatégorie_par_pvl).to_df()
nb_resultats_par_plv
```

```python
#récupération résultat dernier prélévement
query_nb_resultat_dernier_plv = """
WITH 
MAPPED_EDC_RESULTATS AS (
SELECT edc_resultats.*, mapping.categorie
FROM edc_resultats
LEFT JOIN (SELECT DISTINCT cdparametresiseeaux, categorie FROM mapping_categories) AS mapping
ON edc_resultats.cdparametresiseeaux = mapping.cdparametresiseeaux)
,
LAST_PREL AS ( 
SELECT cdreseau, referenceprel, MAX(dateprel) AS dernierdateplv, plvconformitebacterio, plvconformitechimique, 
        plvconformitereferencebact, plvconformitereferencechim
FROM edc_prelevements
WHERE cdreseauamont IS NULL
GROUP BY cdreseau, referenceprel, plvconformitebacterio, plvconformitechimique, 
        plvconformitereferencebact, plvconformitereferencechim
ORDER BY referenceprel, dernierdateplv)
,
TEMP_RESULTS AS ( 
SELECT referenceprel, categorie, referenceanl, cdparametresiseeaux,rqana, limitequal, refqual, valtraduite 
FROM MAPPED_EDC_RESULTATS
ORDER BY referenceprel, categorie )
,
NB_RESULT_PAR_PLV AS (
SELECT referenceprel, COUNT(DISTINCT categorie) AS nbcategorieparplv
FROM TEMP_RESULTS
GROUP BY referenceprel
ORDER BY nbcategorieparplv DESC)
,
NB_RESULT_DERNIER_PLV AS (
SELECT LAST_PREL.* , NB_RESULT_PAR_PLV.nbcategorieparplv
FROM LAST_PREL
LEFT JOIN NB_RESULT_PAR_PLV 
ON LAST_PREL.referenceprel = NB_RESULT_PAR_PLV.referenceprel)

SELECT cdreseau, referenceprel, nbcategorieparplv
FROM NB_RESULT_DERNIER_PLV
ORDER BY nbcategorieparplv
"""

nb_resultat_dernier_plv = con.sql (query_nb_resultat_dernier_plv).to_df()
nb_resultat_dernier_plv
```

```python
#max et min résultats à afficher 
max_nb_resultat = nb_resultat_dernier_plv["nbcategorieparplv"].max()
print(f"nombre maximum d'analyse par catégorie par UDI : {max_nb_resultat}")

max_nb_resultat = nb_resultat_dernier_plv["nbcategorieparplv"].min()
print(f"nombre minimum d'analyse par catégorie par UDI : {max_nb_resultat}")
```
