On va prendre le max de date de prélévement pour afficher le dernier prélévement de chaque UDI

```python
#query du plus récent prélèvement
query_max_prel = """

SELECT CONCAT(referenceprel,'-',dateprel) AS cdplv, referenceprel, dateprel 
FROM (
SELECT referenceprel, MAX(dateprel) AS dateprel
FROM edc_prelevements
GROUP BY referenceprel)

"""
dernier_plv =con.sql(query_max_prel).fetchdf()
dernier_plv

#nbreseau = dernier_plv['referenceprel'].nunique()
#nbreseau
```

```python
#creation de clé unique dans table edc_prélèvement (cdplv = referenceprel-dateprel)
query_unique_plv = """
SELECT CONCAT(referenceprel,'-',dateprel) AS cdplv, cdreseau,inseecommuneprinc, nomcommuneprinc, cdreseauamont, nomreseauamont, referenceprel, dateprel
FROM edc_prelevements
WHERE referenceprel IN (SELECT referenceprel
    FROM edc_prelevements
    GROUP BY referenceprel
    HAVING COUNT(referenceprel) > 1)
ORDER BY referenceprel, dateprel
"""
unique_plv =con.sql(query_unique_plv).fetchdf()
unique_plv
```

```python
#max prélèvement unique avec autres colonnes
query_unique_max_plv_allcol = """
WITH unique_recent_plv AS (
SELECT CONCAT(referenceprel,'-',dateprel) AS cdplv, referenceprel, dateprel 
FROM (
SELECT referenceprel, MAX(dateprel) AS dateprel
FROM edc_prelevements
GROUP BY referenceprel)) ,

edc_prelevements_cdplv AS (
SELECT CONCAT(referenceprel,'-',dateprel) AS cdplv, cdreseau,inseecommuneprinc, nomcommuneprinc, cdreseauamont, nomreseauamont, referenceprel, dateprel
FROM edc_prelevements)



SELECT  cdplv
FROM edc_prelevements_cdplv
WHERE cdplv IN (SELECT cdplv FROM unique_recent_plv )


"""
edc_plv_unique = con.sql(query_unique_max_plv_allcol).fetchdf()
edc_plv_unique

```

```python
query_join = """
SELECT edc_prelevements.cdreseau, edc_prelevements.referenceprel, edc_prelevements.de_ingestion_date,edc_prelevements.inseecommuneprinc, edc_prelevements.cdreseauamont, edc_prelevements.nomreseauamont, result_per_prel.nbresult
FROM edc_prelevements
LEFT JOIN (SELECT referenceprel, COUNT(cdparametre) AS nbresult 
            FROM edc_resultats 
            GROUP BY referenceprel) AS result_per_prel
ON edc_prelevements.referenceprel = result_per_prel.referenceprel
WHERE cdreseau = '001000003'
ORDER BY edc_prelevements.cdreseau
"""

resultat_par_plv = con.sql(query_join).df()
resultat_par_plv
```
